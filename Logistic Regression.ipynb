```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

```

```python
X, Y = make_blobs(n_samples=100, centers=2, n_features=1)
```

```python
def sigmoid(x, y, w, b):
    val = (w*x) + b
    sig = 1 / (1 + (np.exp(-val)))
    #print("Exponent", val, sig)
    return sig

def cost_func(x, y, w, b):
    y_pred = sigmoid(x, y, w, b)
    cost = -( (y*np.log(y_pred)) + ((1-y)*np.log(1-y_pred)) )
#     print("y_pred", y_pred)
#     print("1st part", (y*np.log(y_pred)))
#     print("2nd part", (1-y)*np.log(1-y_pred))
#     print("Log", np.log(y_pred), np.log(1-y_pred))
#     print("X, Y, W, B", x, y, w, b)
#     print("cost", cost)
    return cost

def gradient(x, y, w, b):
    y_pred = sigmoid(x, y, w, b)
    return (y_pred-y)*x, (y_pred-y)

def logistic_regression(X, Y, w, b, L=0.01):
    total_cost = []
    for ind,x in enumerate(X):
        y = Y[ind]
        cost = cost_func(x, y, w, b)
        total_cost.append(cost[0])
        #print(ind, "Cost = ", cost)
        dw, db = gradient(x, y, w, b)
        w = w - (L*dw)
        b = b - (L*db)
        #print("==============================")
    total_cost = sum(total_cost) / len(X)
    return w, b, total_cost
        
    
    
```

```python
w = b = 0
n_iter = 10000



for i in range(n_iter):
    w,b,total_cost = logistic_regression(X, Y, w, b, L=0.0003)
    print(i, "Cost = ", total_cost)
```
