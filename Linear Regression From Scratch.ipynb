```python
def hypothesis(m, b, x):
    return ((m*x)+b)

def cost_func(data, m , b):
    sumi = 0
    for i in data:
        y_pred = hypothesis(m, b, i[0])
        sumi += (y_pred-i[1])**2
    return sumi/len(data)

def gradi(data, m , b):
    grad_b_sumi = grad_m_sumi = 0
    predicted = []
    for i in data:
        y_pred = hypothesis(m, b, i[0])
        #print("data", i)
        #print("predicted ", y_pred)
        predicted.append(y_pred)
        grad_b_sumi += (y_pred-i[1])
        grad_m_sumi += (y_pred-i[1]) * i[0]
        #print("grad_b_sumi", "grad_m_sumi", grad_b_sumi, grad_m_sumi)
    grad_b = (2/len(data)) * grad_b_sumi
    grad_m = (2/len(data)) * grad_m_sumi
    #print("grad_b, grad_m", grad_b, grad_m)
    return grad_b, grad_m, predicted

def plot_lin(data, predicted):
    plt.figure(figsize=(4, 3))
    ax = plt.axes()
    ax.scatter([x[0] for x in data], [x[1] for x in data])
    ax.plot([x[0] for x in data], predicted)

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    plt.show()


def lin_reg(data):
    m = b = 0
    l = 0.002
    for i in range(1001):
        cost = cost_func(data, m, b)
        print(i, "cost = ", cost)
        grad = gradi(data, m , b)
        if (i % 10) == 0:
            plot_lin(data, grad[2])
        #print(grad)
        b = b - (l * (grad[0]))
        m = m - (l * (grad[1]))
    print("line =", m, b)
        

        
        
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=2)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=3)
```

```python
import numpy as np


class LinearRegression:

    def __init__(self, lr = 0.005, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            y_pred = np.dot(X, self.weights) + self.bias
            #print(y_pred)
            dw = (1/n_samples) * np.dot(X.T, (y_pred-y))
            db = (1/n_samples) * np.sum(y_pred-y)
            #print("dw", dw)
            #print("db", db)
            self.weights = self.weights - self.lr * dw
            self.bias = self.bias - self.lr * db

    def predict(self, X):
        y_pred = np.dot(X, self.weights) + self.bias
        return y_pred
```
